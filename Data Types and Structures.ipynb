2.Explain the difference between mutable and immutable data types with examples.
 - Mutable data types can be changed after creation, while immutable data types cannot.

∆ Mutable data type.

•Can be changed after creation.

•Ideal for dynamic data.

 ∆ Immutable data types.

•Cannot be changed after creation.

•Values or properties cannot be modified directly.

5.Why might you use a set instead of a list in Python?
 -A set in Python is preferred over a list in scenarios prioritizing uniqueness and efficient membership testing. Sets, by definition, store only unique elements, automatically eliminating duplicates. This feature proves valuable when dealing with data where redundancy is undesirable. Additionally, sets utilize hashing for element storage, enabling near-constant time complexity, O(1), for membership checks (i.e., determining if an element exists within the set). In contrast, lists require linear search, O(n), for membership testing, where n is the number of elements in the list.

Consider the following scenarios:

•Removing Duplicates: If the primary goal is to eliminate duplicate entries from a collection of items, converting the list to a set achieves this efficiently.

```python
    my_list = [1, 2, 2, 3, 4, 4, 5]
    my_set = set(my_list)
    print(my_set)  # Output: {1, 2, 3, 4, 5}
```

•Membership Testing: When frequently checking for the presence of an element within a collection, a set outperforms a list due to its faster lookup time.

```python
    my_set = {1, 2, 3, 4, 5}
    print(2 in my_set)  # Output: True
```

•Mathematical Set Operations: Sets are optimized for operations like union, intersection, and difference, which are commonly used in mathematical and data analysis contexts.

```python
    set1 = {1, 2, 3}
    set2 = {3, 4, 5}
    print(set1.union(set2))  # Output: {1, 2, 3, 4, 5}
```

•Order Irrelevance: If the order of elements is not crucial, a set provides the benefits of uniqueness and efficient operations without the overhead of maintaining order.

However, lists are more appropriate when the order of elements matters or when duplicate values are permissible.
